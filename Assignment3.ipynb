**OPTIMISATION**
Find the global minimum point and value for the function
-do manual calculation for two iterations
-find the optimal solution using python
3x^2+5e^-y+10

```python
import numpy as np
import math
x=np.random.uniform(-10,10)
y=np.random.uniform(-10,10)
eta=0.1
iter_max=30
iter=1
while iter<=iter_max:
  g_x= 6*x
  g_y= -5*(math.exp(-y))
  d_x=-eta*g_x
  d_y=-eta*g_y
  x=x+d_x
  y=y+d_y
  iter=iter+1
obj=3*x*x+5*math.exp(-y)+10
print("Minimum =",obj,"at X=",x,"Y=",y)
```
